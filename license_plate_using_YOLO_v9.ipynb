#### Open data.yml file in any text editor and change the test, train and val path with these:

test: /content/gdrive/MyDrive/License_Plate/dataset/ANPR2-8/test/images

train: /content/gdrive/MyDrive/License_Plate/dataset/ANPR2-8/train/images

val: /content/gdrive/MyDrive/License_Plate/dataset/ANPR2-8/valid/images


You can do it using automation uuing the following code

```python



# Read the file content
with open('/content/gdrive/MyDrive/License_Plate/yolov9/data.yaml', 'r') as file:
    lines = file.readlines()

# Modify the specific lines
modified_lines = []
for line in lines:
    if 'test: ../test/images' in line:
        modified_lines.append('test: /content/gdrive/MyDrive/License_Plate/dataset/ANPR2-8/test/images\n')
    elif 'train: ANPR2-8/train/images' in line:
        modified_lines.append('train: /content/gdrive/MyDrive/License_Plate/dataset/ANPR2-8/train/images\n')
    elif 'val: ANPR2-8/valid/images' in line:
        modified_lines.append('val: /content/gdrive/MyDrive/License_Plate/dataset/ANPR2-8/valid/images\n')
    else:
        modified_lines.append(line)

# Write the modified content back to the file
with open('/content/gdrive/MyDrive/License_Plate/yolov9/data.yaml', 'w') as file:
    file.writelines(modified_lines)

# Display the modified file content
with open('/content/gdrive/MyDrive/License_Plate/yolov9/data.yaml', 'r') as file:
    print(file.read())
```

```python
!cp /content/gdrive/MyDrive/License_Plate/yolov9/models/detect/yolov9-e.yaml /content/gdrive/MyDrive/License_Plate/yolov9/models/detect/yolov9-e_custom.yaml

# Read the file content
with open('/content/gdrive/MyDrive/License_Plate/yolov9/models/detect/yolov9-e_custom.yaml', 'r') as file:
    lines = file.readlines()

# Modify the specific line
modified_lines = []
for line in lines:
    if 'nc: 80' in line:
        modified_lines.append(line.replace('nc: 80', 'nc: 1'))
    else:
        modified_lines.append(line)

# Write the modified content back to the file
with open('/content/gdrive/MyDrive/License_Plate/yolov9/models/detect/yolov9-e_custom.yaml', 'w') as file:
    file.writelines(modified_lines)

# Display the modified file content
with open('/content/gdrive/MyDrive/License_Plate/yolov9/models/detect/yolov9-e_custom.yaml', 'r') as file:
    print(file.read())
```
